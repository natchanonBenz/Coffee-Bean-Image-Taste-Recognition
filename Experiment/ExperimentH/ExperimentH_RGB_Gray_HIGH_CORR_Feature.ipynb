```python
import numpy as np 
import pandas as pd 
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
df_Mobile = pd.read_excel('C:/Users/U5988116/Desktop/SeniorProject/Data/Database/Main/coffee_bean_Mobile_v2.xlsx')  
df_Mobile
```

```python
# Train Mobile
df= df_Mobile
d1= df.iloc[:, 1:3]
d2= df.iloc[:,4:5]
d3= df.iloc[:,6:10]
d4 = df.iloc[:,11:12]
hi_df = pd.concat([d1,d2,d3,d4], axis=1)
hi_df
```

```python
# (Mobile)
df_Mobile= hi_df
BAL_1 = df_Mobile.iloc[0:10,0:7] #BAL
BPL_1 = df_Mobile.iloc[10:20,0:7] #BPL
CM_2= df_Mobile.iloc[20:30,0:7] #CM
BAM_3 = df_Mobile.iloc[50:60,0:7]#BAM
BPM_3 = df_Mobile.iloc[60:70,0:7]#BPM
BAD_4 = df_Mobile.iloc[90:100,0:7]#BAD
PD_4 = df_Mobile.iloc[100:110,0:7]#PD
PM_2= df_Mobile.iloc[30:40,0:7]#PM
BAL_1
```

```python
G1= pd.concat([BAL_1, BPL_1])
feature_G1 = G1
G2= pd.concat([CM_2,PM_2])
feature_G2 = G2
G3=pd.concat([BAM_3,BPM_3])
feature_G3 = G3
G4=pd.concat([BAD_4,PD_4])
feature_G4 = G4
```

```python
G1_target = pd.concat([Target_BAL_1, Target_BPL_1])
G2_target = pd.concat([Target_CM_2,Target_PM_2])
G3_target = pd.concat([Target_BAM_3,Target_BPM_3])
G4_target = pd.concat([Target_BAD_4,Target_PD_4])
```

```python
# 80% training and 20% test (Iphone)
fea_train_G1, fea_test_G1, tar_train_G1, tar_test_G1 = train_test_split(feature_G1, G1_target,test_size=0.2, random_state=1)
fea_train_G2, fea_test_G2, tar_train_G2, tar_test_G2 = train_test_split(feature_G2, G2_target,test_size=0.2, random_state=1)
fea_train_G3, fea_test_G3, tar_train_G3, tar_test_G3 = train_test_split(feature_G3, G3_target, test_size=0.2, random_state=1)
fea_train_G4, fea_test_G4, tar_train_G4, tar_test_G4 = train_test_split(feature_G4, G4_target,test_size=0.2, random_state=1)
```

```python
df = hi_df
Target_BAL_1_AN = df.iloc[0:10,7:8]#BAL
Target_BPL_1_AN = df.iloc[10:20,7:8]#BPL
Target_CM_2_AN= df.iloc[20:30,7:8]#CM
Target_BAM_3_AN = df.iloc[50:60,7:8]#BAM
Target_BPM_3_AN = df.iloc[60:70,7:8]#BPM
Target_BAD_4_AN = df.iloc[90:100,7:8]#BAD
Target_PD_4_AN = df.iloc[100:110,7:8]#PD
Target_PM_2_AN= df.iloc[30:40,7:8]#PM
```

```python
G1_target_AN = pd.concat([Target_BAL_1_AN, Target_BPL_1_AN])
G2_target_AN = pd.concat([Target_CM_2_AN,Target_PM_2_AN])
G3_target_AN = pd.concat([Target_BAM_3_AN,Target_BPM_3_AN])
G4_target_AN = pd.concat([Target_BAD_4_AN,Target_PD_4_AN])

```

```python
# 80% training and 20% test (Android)
fea_train_G1_AN, fea_test_G1_AN, tar_train_G1_AN, tar_test_G1_AN = train_test_split(feature_G1_AN, G1_target_AN,test_size=0.2, random_state=1)
fea_train_G2_AN, fea_test_G2_AN, tar_train_G2_AN, tar_test_G2_AN = train_test_split(feature_G2_AN, G2_target_AN,test_size=0.2, random_state=1)
fea_train_G3_AN, fea_test_G3_AN, tar_train_G3_AN, tar_test_G3_AN = train_test_split(feature_G3_AN, G3_target_AN, test_size=0.2, random_state=1)
fea_train_G4_AN, fea_test_G4_AN, tar_train_G4_AN, tar_test_G4_AN = train_test_split(feature_G4_AN, G4_target_AN,test_size=0.2, random_state=1)
tar_test_G1_AN
```

```python
main_feature_train = np.concatenate((  fea_train_G1,fea_train_G1_AN,fea_train_G2,fea_train_G2_AN,fea_train_G3,fea_train_G3_AN,fea_train_G4,fea_train_G4_AN), axis=0)
main_target_train=np.concatenate(( tar_train_G1,tar_train_G1_AN,tar_train_G2,tar_train_G2_AN,tar_train_G3,tar_train_G3_AN,tar_train_G4,tar_train_G4_AN), axis=0)
main_feature_test = np.concatenate(( fea_test_G1,fea_test_G1_AN, fea_test_G2,fea_test_G2_AN, fea_test_G3,fea_test_G3_AN, fea_test_G4,fea_test_G4_AN), axis=0)
main_target_test=np.concatenate(( tar_test_G1,tar_test_G1_AN,tar_test_G2,tar_test_G2_AN,tar_test_G3,tar_test_G3_AN,tar_test_G4,tar_test_G4_AN), axis=0)
```

```python
dTree = DecisionTreeClassifier(max_depth = 30, min_samples_leaf = 4, max_features = 7)
dTree.fit(main_feature_train,main_target_train)
```

```python
Mobile_test_model = dTree.predict(main_feature_test)
Mobile_test_model 
```

```python

print("\nDesicion Tree")

print("\nEvaluation of DTree Model : \n", classification_report(main_target_test,Mobile_test_model ))
```
